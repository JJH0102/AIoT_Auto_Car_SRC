```python
# 선형 회귀

from pop import AI
import matplotlib.pyplot as plt

LR = AI.Linear_Regression()   # 선형회귀 객체 생성

# 입력 데이터 (키)
LR.X_data = [[173], [171], [162], [181], [157], [169], [177], [159], [182]]

# 정답 데이터 (발크기)
LR.Y_data = [[270], [275], [245], [280], [230], [265], [270], [250], [275]]

plt.plot(LR.X_data, LR.Y_data)
plt.show()
```

```python
plt.plot(LR.X_data, LR.Y_data, 'ro')
plt.show()
```

```python
LR.train()
```

```python
# 데이터 검증(훈련데이터로 예측값 출력)

pred = LR.run()
```

```python
plt.plot(LR.X_data, pred, 'ro')
plt.show()
```

```python
LR.train(times = 1000, print_every = 100)
```

```python
pred = LR.run()
```

```python
X_pred = ([150], [160], [170], [180], [190])
pred = LR.run(X_pred)
```

```python
plt.plot(X_pred, pred, 'ro')
plt.show()
```

```python
# 로지스틱 회귀

from pop import AI

LR = AI.Logistic_Regression()

# 키 데이터
LR.X_data = [[173], [171], [162], [187], [157], [169], [177], [159], [182]]

# 성별 데이터(1 = 여자, 0 = 남자)
LR.Y_data = [[0], [1], [1], [0], [1], [0], [0], [1], [0]]

LR.train()
pred = LR.run()
```

```python
LR.train(times = 10000, print_every = 1000)
```

```python
plt.plot(LR.X_data, pred, 'ro')
plt.show()
```
