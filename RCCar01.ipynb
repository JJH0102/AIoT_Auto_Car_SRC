```python
import time
from pop import Pilot

Car = Pilot.AutoCar()

# 카메라 좌우 제어
Car.camPan(0) # 좌로 이동
time.sleep(1)
Car.camPan(180) # 우로 이동
time.sleep(1)

# 카메라 상하 제어
Car.camTilt(180) # 상으로 이동
time.sleep(1)
Car.camTilt(0) # 하로 이동
time.sleep(1)
```
